# Admin Pages Audit for Payroll, Documents, Notifications, Reports, and Settings

This notebook inspects the admin pages and backend endpoints for payroll, documents, notifications, reports, and settings. It also parses the attached requirements PDF for expected functionality.

```python
import os
from pathlib import Path
from pprint import pprint

ROOT = Path('d:/projs/proj-fin')
SRC = ROOT / 'conexra-frontend' / 'src'
BACKEND = ROOT / 'backend' / 'src'
PDF_PATH = ROOT / 'Conexra_HR_Administration_Operations_Platform_Requirements.pdf'
print('ROOT', ROOT)
print('SRC', SRC)
print('BACKEND', BACKEND)
print('PDF exists:', PDF_PATH.exists())
```

```python
try:
    from PyPDF2 import PdfReader
except ImportError:
    PdfReader = None
    print('PyPDF2 unavailable; PDF parsing may fail.')

def extract_pdf_text(path, max_pages=None):
    if PdfReader is None:
        return 'PyPDF2 not installed.'
    reader = PdfReader(str(path))
    text = []
    for i, page in enumerate(reader.pages):
        if max_pages is not None and i >= max_pages:
            break
        text.append(page.extract_text() or '')
    return '

'.join(text)

pdf_text = extract_pdf_text(PDF_PATH, max_pages=8)
print(pdf_text[:2000])
```

```python
def find_files(root, patterns):
    results = []
    for pattern in patterns:
        results.extend(sorted(root.rglob(pattern)))
    return results

frontend_pages = find_files(SRC, ['pages/*.jsx', 'pages/*.js', 'pages/**/*.jsx', 'pages/**/*.js'])
backend_files = find_files(BACKEND, ['controllers/*.js', 'routes/*.js', 'models/*.js'])
print('Frontend page files:')
for path in sorted(frontend_pages):
    print('-', path.relative_to(ROOT))
print('
Backend related files:')
for path in sorted(backend_files):
    print('-', path.relative_to(ROOT))
```

```python
admin_pages = ['Payroll.jsx', 'Documents.jsx', 'Notifications.jsx', 'Reports.jsx', 'Settings.jsx']
for page in admin_pages:
    p = SRC / 'pages' / page
    print(page, 'exists' if p.exists() else 'missing', p)

backend_targets = ['payrollController.js', 'documentController.js', 'notificationController.js', 'reportController.js', 'settingsController.js', 'payrollRoutes.js', 'documentRoutes.js', 'notificationRoutes.js', 'reportRoutes.js', 'settingsRoutes.js']
for fname in backend_targets:
    p = BACKEND / 'controllers' / fname if 'Controller' in fname else BACKEND / 'routes' / fname
    print(fname, 'exists' if p.exists() else 'missing', p)
```

```python
def read_text(path, max_lines=200):
    if not path.exists():
        return f'Missing file: {path}'
    text = path.read_text(encoding='utf-8', errors='ignore').splitlines()
    return '
'.join(text[:max_lines])

for page in admin_pages:
    p = SRC / 'pages' / page
    print('---', page, '---')
    print(read_text(p, max_lines=120))
    print()
```

```python
controllers = ['payrollController.js', 'documentController.js', 'notificationController.js', 'reportController.js', 'settingsController.js']
for c in controllers:
    p = BACKEND / 'controllers' / c
    print('---', c, '---')
    print(read_text(p, max_lines=160))
    print()
```

```python
routes = ['payrollRoutes.js', 'documentRoutes.js', 'notificationRoutes.js', 'reportRoutes.js', 'settingsRoutes.js']
for r in routes:
    p = BACKEND / 'routes' / r
    print('---', r, '---')
    print(read_text(p, max_lines=120))
    print()
```
